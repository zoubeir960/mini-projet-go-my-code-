# part C:  machine learning-unsupervised-Kmeans: 

```python
#Importing our clustering algorithm: KMeans
from sklearn.cluster import KMeans  

kmeans=KMeans(n_clusters=6, random_state=0)  #Cluster our data by choosing 6 as number of clusters
kmeans.fit(data)
```

```python
#K-means Labels
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
#K-means Labels
labels=pd.DataFrame(kmeans.labels_)
labels.value_counts()
```

```python
#K-means Centroids
kmeans.predict(data)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
# Partitional Clustering (K- Means) Using Elbow Method
```

```python
sum_of_squared_distances = []
k = range(1,15)
for k in k:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(range(1, 15), sum_of_squared_distances, 'bx-')
plt.xlabel('Number of Cluster')
plt.ylabel('Sum of Squared Error')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
#Build Model with 2 Cluster
kmeans=KMeans(n_clusters=2, random_state=0)  
kmeans.fit(data)
```
